```python
MAXNUM = 4
MAXHEIGHT = 130
more = True
cnt = 0
while more:
  height = float(input("키는 ? "))
  if height < MAXHEIGHT:
    cnt += 1
    print('들어가세요.', '%d명' % cnt)
  else:
    print('커서 못 들어갑니다.')
  if cnt == MAXNUM:
    more = False
else:
  print('%d명 모두 찼습니다. 다음 번에 이용하세요.' % cnt)
```

```python
n = int(input('소수(prime number)인지를 판별한 2 이상의 정수 입력 >> '))
cnt = 2
for i in range(2, n // 2 + 1):
  if n % i == 0:
    cnt += 1
    break
if cnt == 2:
  print('정수 %d는 소수입니다.' % n)
else:
  print('정수 %d는 소수가 아닙니다.' % n)
```
